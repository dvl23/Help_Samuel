```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
im = cv2.imread('S1_0.01.jpg')
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(im)
```

```python
np.average(im[:,:,0])# average of red channel in this entire image
```

```python
np.average(im[:, :,1])# average of green channel in this entire image
```

```python
np.average(im[:,:,2])#average of blue channel in this entire image
```

```python
np.average(im, axis = (0,1))#sanity test to check if the pixels values match
```
